## Exercício 11: Crie um menu onde o usuário pode escolher entre somar, subtrair, multiplicar e dividir dois números.

```python
menu = True
while(menu):

  numero1 = int(input("digite o primeiro numero: "))
  numero2 = int(input("digite o segundo numero: "))

  print("""
  1) somar
  2) subtrair
  3) multiplicar
  4) dividir
  5) sair
  """)

  opcao = int(input("digite uma opção: "))

  if(opcao == 1):
    soma = numero1 + numero2
    print(f" a soma entre o {numero1} e o {numero2} é: {soma}")
  elif(opcao == 2):
    subtrair = numero1 - numero2
    print(f" a soma entre o {numero1} e o {numero2} é: {subtrair}")
  elif(opcao == 3):
    multiplicar = numero1 * numero2
    print(f" a soma entre o {numero1} e o {numero2} é: {multiplicar}")
  elif(opcao == 4):
    dividir = numero1 / numero2
    print(f" a soma entre o {numero1} e o {numero2} é: {dividir}")
  elif(opcao == 5):
    menu = False
  else:
    print("digite um numero valido como opção")
```


## Exercício 12: Crie um menu para converter temperaturas entre Celsius, Fahrenheit e Kelvin.

```python
while True:
    print("\n=== Conversor de Temperaturas ===")
    print("1 - Celsius para Fahrenheit e Kelvin")
    print("2 - Fahrenheit para Celsius e Kelvin")
    print("3 - Kelvin para Celsius e Fahrenheit")
    print("0 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "0":
        print("Vlw")
        break

    temp = float(input("Digite a temperatura: "))

    if opcao == "1":
        fahrenheit = (temp * 9/5) + 32
        kelvin = temp + 273.15
        print(f"{temp}°C = {fahrenheit:.2f}°F")
        print(f"{temp}°C = {kelvin:.2f}K")
    elif opcao == "2":
        celsius = (temp - 32) * 5/9
        kelvin = celsius + 273.15
        print(f"{temp}°F = {celsius:.2f}°C")
        print(f"{temp}°F = {kelvin:.2f}K")
    elif opcao == "3":
        celsius = temp - 273.15
        fahrenheit = (celsius * 9/5) + 32
        print(f"{temp}K = {celsius:.2f}°C")
        print(f"{temp}K = {fahrenheit:.2f}°F")
    else:
        print("Opção inválida! Tente novamente.")
```

## Exercício 13: Crie um programa que traduza números de 1 a 5 para palavras (1 = um, 2 = dois...).

```python
numero = int(input("Digite um número de 1 a 5: "))

match numero:
    case 1:
        print("Um")
    case 2:
        print("Dois")
    case 3:
        print("Três")
    case 4:
        print("Quatro")
    case 5:
        print("Cinco")
    case _:
        print("Número fora do intervalo!")
```

## Exercício 46: Crie um dicionário com nomes de alunos e suas notas e exiba a nota de um aluno específico.

```python
alunos = [
    {"nome":"Ana", "idade":20, "nota":8.5},
    {"nome":"Bruno", "idade":22, "nota":7.0},
    {"nome": "Carlos", "idade": 21, "nota": 6.5},
    {"nome": "Daniela", "idade": 20, "nota": 9.0},
    {"nome": "Eduarda", "idade": 23, "nota": 5.5}
]

def mostrar_nota(nome):
  for i in alunos:
    if nome == i["nome"]:
      print(i["nota"])

aluno = input("Digite o nome de um aluno:")

mostrar_nota(aluno)
```
